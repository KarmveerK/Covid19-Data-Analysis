## Importing necessary packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
print("modules are imported")
```

```python
corona_dataset_csv = pd.read_csv(r"C:\Users\Karmveer\Desktop\datasets\covid19_Confirmed_dataset.csv")
corona_dataset_csv.head()
```

```python
corona_dataset_csv.shape
```

## Removing the useless columns from the dataset

```python
useless_cols=['Lat','Long']
corona_dataset_csv.drop(useless_cols, axis=1, inplace=True)
```

```python
corona_dataset_csv.head()
```

## Aggregating the data for each country as some data is divided into provinces

```python
corona_dataset_aggregated = corona_dataset_csv.groupby("Country/Region").sum()
```

```python
corona_dataset_aggregated.head()
```

```python
corona_dataset_aggregated.loc['China',].plot()
corona_dataset_aggregated.loc['India',].plot()
corona_dataset_aggregated.loc['Italy',].plot()
plt.legend()
```

## Finding the first derivative of the data to find the rate of infection

```python
corona_dataset_aggregated.loc['China',].diff().plot()
```

```python
corona_dataset_aggregated.loc['China',].diff().max()
```

```python
corona_dataset_aggregated.loc['India',].diff().plot()
```

```python
corona_dataset_aggregated.loc['India',].diff().max()
```

## Making a list of maximum infection rates for all countries

```python
countries = list(corona_dataset_aggregated.index)
max_infection_rates = list()
for country in countries:
    max_infection_rates.append(corona_dataset_aggregated.loc[country].diff().max())
max_infection_rates
```

```python
corona_dataset_aggregated['max_infection_rates'] = max_infection_rates
corona_dataset_aggregated.head()
```

```python
corona_data = corona_dataset_aggregated.loc[:,['max_infection_rates']]
corona_data.head()
```

```python
world_happiness_report = pd.read_csv(r"C:\Users\Karmveer\Desktop\datasets\worldwide_happiness_report.csv")
```

```python
world_happiness_report.head()
```

```python
columns_to_dropped = ['Overall rank','Score','Generosity','Perceptions of corruption']
world_happiness_report.drop(columns_to_dropped,axis=1 , inplace=True)
```

```python
world_happiness_report.head()
```

## Setting the index of data as country names so the we can add the data to covid data 

```python
world_happiness_report.set_index(['Country or region'],inplace=True)
```

```python
world_happiness_report.head()
```

## Joining the data as inner as shape of data is different in both cases

```python
data = corona_data.join(world_happiness_report, how='inner')
```
